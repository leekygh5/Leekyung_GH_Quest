### Pandas를 이용한 데이터 불러오기 & 정리

```python
import pandas as pd
import numpy as np
```

```python
#CCTV_Seoul 불러오기
CCTV_Seoul = pd.read_csv('C:\\Users\\LEE KO\\Desktop\\CCTV_in_Seoul.csv', encoding = 'utf-8') 
CCTV_Seoul.head()
```

# 1.
### 데이터셋’ 폴더의 CCTV_in_Seoul.csv, population_in_Seoul.csv를 이용, 
### pandas / numpy를 통해 아래와 같은 형태로 데이터를 정리해 볼 것. 
### '외국인비율’ = ‘외국인’ / ‘인구수’ * 100, ‘고령자비율’ = ‘고령자’ / ‘인구수’ * 100
### ‘오차’ – l소계 – 회귀직선(인구수)l , 오차는 np.abs를 활용


```python
import matplotlib.pyplot as plt 
%matplotlib inline
import platform                

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus']= False

if platform.system() == 'Darwin': #맥os 사용자
    rc('font', family = 'AppleGothic')
elif platform.system() == 'Windows':#윈도우 사용자
    path = 'c:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)

    
    
#인구수-소계 일차함수 그래프 (실습파일 matplotlib 부분에 있던 부분)
fp1 = np.polyfit(data_result['인구수'], data_result['소계'], 1)
f1 = np.poly1d(fp1) 
fx = np.linspace(100000, 700000, 100)
    
#새로운 칼럼 만들기
data_result['외국인비율'] = (data_result['외국인'] / data_result['인구수']*100)
data_result['고령자비율'] = (data_result['고령자']/ data_result['인구수']*100)
data_result['CCTV비율'] = (data_result['소계']/ data_result['인구수']*100)
data_result['오차'] = np.abs(data_result['소계']-f1(data_result['인구수']))
                                                      
data_result=data_result.sort_values(by='오차', ascending=False)
data_result.head()
```

# 2
### (1)에서 정리한 데이터를 ‘오차’ 컬럼을 기준으로 색깔을 입혀서 시각화 해보고
### 이를 통해 무엇을 파악할 수 있는지 주석으로 간략하게 1-2줄로 서술 (plt.colormap, plt.text 로 컬러맵과 이름 추가)

```python
#scatterplot 데이터에 색깔 넣기
colormap2=data_result['오차']
name=data_result.index

plt.figure(figsize=(8,6))
plt.plot(fx, f1(fx), ls='dashed', lw=3, c= 'g')
plt.scatter(data_result['인구수'], data_result['소계'], c = colormap2) 
plt.colorbar()
plt.xlabel('인구수')
plt.ylabel('CCTV소계')
plt.grid()

#이름 넣기 (오차 큰 순서로)
for i in range(10):
    plt.text(data_result['인구수'][i], data_result['소계'][i], name[i], fontsize = 10)

plt.show()

#파악할 수 있는 점: 대체로 인구수가 많은 지역구일수록 CCTV가 많이 설치되어 있다. 
#하지만 이름이 표시된 몇 개의 지역구의 경우 인구수-CCTV 소계 간의 일차 비례식으로 계산되지 않는 오차를 보인다.
```
