```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns  
from sklearn import preprocessing
from sklearn.preprocessing import scale
```

```python
dataset = pd.read_csv('2017.csv')
```

```python
#결측치 없음 확인
dataset.info()
```

```python
#모든 변수에 대해 널값 없음 확인
dataset.isnull().sum()
```

```python
#현재 칼럼 이름이 긴것들이 많으므로 편의상 바꿔주기
simple_dataset=dataset.rename(columns={'Economy..GDP.per.Capita.':'GDP', 'Health..Life.Expectancy.':'Life.Expectancy', 'Trust..Government.Corruption.':'Government.Trust'})
```

```python
#이름 바뀐거 확인
simple_dataset.head(10)
```

```python
#kmeans clustering으로 Happiness.Score 이랑 GDP 해보려고 함. 해당 칼럼만 남기기
X= simple_dataset.iloc[:, [5,2]].values
X
```

```python
#정규화.standardization
scaler = preprocessing.StandardScaler()
scaled_data = scaler.fit_transform(X)
scaled_data= pd.DataFrame(scaled_data, columns=['GDP','Happiness.Score'])
scaled_data.head(10)
```

```python
# 모델 만들기, kmeans clustering 위한 적절한 k 값 구하기
# score 값 리스트 생성 후 적절한 k값 찾기 
from sklearn.cluster import KMeans
score=[]
```

```python
for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(X) 
    score.append(kmeans.inertia_)
```

```python
score
```

```python
#최적의 k 값 찾기위해 시각화 --> 그래프 보면 k=3이 적절
plt.figure(figsize=(8, 8))
plt.plot(range(1,11), score)
plt.title('The Elbow Method')
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
# k=3으로 모델 만들기 
kmeansmodel = KMeans(n_clusters= 3, init='k-means++', random_state=0)
y_kmeans= kmeansmodel.fit_predict(X)
```

```python
#클러스터 색깔 다르게 시각화 
plt.figure(figsize=(12, 12))
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1') 
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.title('Clusters of Countries')
plt.xlabel('GDP per Capita') 
plt.ylabel('Happiness Score') 
plt.legend()
plt.show()
```

#### 대체적으로 GDP와 행복지수는 비례 관계를 보이는 것 같은데 clustering은 3개의 집단으로 나눌때가 가장 설명력이 좋게 나왔다. 
#### 이 결과를 참고하여 (높은 수준의 GDP & 높은 행복지수), (중간 수준의 GDP & 중간 정도의 행복지수), (낮은 수준의 GDP & 낮은 정도의 행복지수) 이렇게 국가군을 나눌 수 있다.
